# Formulas estadisticas en Python
 
* x.count( ) =>> Contar
* x.size( ) =>> Tamaño 
* x.mean( ) =>> Proemdio
* x.median( ) =>> Mediana 
* x.mode( ) =>> Moda
* x.std( ) =>> Desviacion tipica
* x.var( ) =>> Varianza
* x.quantile( ) =>> Cuartile
* x.min( ) =>> minimo
* x.max( ) =>> Maximo
* x.describe( ) =>> Descritivo
* x.cumsum( ) =>> suma acumulada
* x.cummax( ) =>> max acumulada
* x.cummin( ) =>> min acumulada
* x.cumprod( ) =>> productoria

```python
aguacates.head()
```

```python
def Descriptivo(x):
    media = []
    mediana = []
    varianza = []
    desviacion = []
    maximo = []
    minimo = []
    conteo = []
    moda = []
    promedio = []
    descriptivo = []
    sumaacu = []
    maxacu = []
    minacu = []
    productoria = []
    media.append(x.mean())
    mediana.append(x.quantile(0.5))
    varianza.append(x.var())
    desviacion.append(x.std())
    maximo.append(x.max())
    minimo.append(x.min())
    conteo.append(x.count())
    moda.append(x.mode())
    promedio.append(x.mean())
    descriptivo.append(x.describe())
    sumaacu.append(x.cumsum()) 
    maxacu.append(x.cummax())
    minacu.append(x.cummin())
    productoria.append(x.cumprod())
    resultados = pd.DataFrame({'media':media, 'mediana':mediana, 'varianza':varianza, 'desviacion':desviacion,
                  'maximo':maximo, 'minimo':minimo, 'conteo':conteo, 'moda':moda, 'promedio':promedio,
                              'descriptivo':descriptivo, 'sumaacu':sumaacu, 'maxacu':maxacu, 'minacu':minacu, 
                               'productoria':productoria})
    return resultados
```

```python
def Descriptivo(x):
    variable = []
    media = []
    mediana = []
    varianza = []
    desviacion = []
    maximo = []
    minimo = []
    conteo = []
    moda = []
    promedio = []
    descriptivo = []
    sumaacu = []
    maxacu = []
    minacu = []
    productoria = []
    for i in range(len(x.columns)):
        variable.append(x.columns[i])
        media.append(x[x.columns[i]].mean())
        mediana.append(x[x.columns[i]].quantile(0.5))
        varianza.append(x[x.columns[i]].var())
        desviacion.append(x[x.columns[i]].std())
        maximo.append(x[x.columns[i]].max())
        minimo.append(x[x.columns[i]].min())
        conteo.append(x[x.columns[i]].count())
        moda.append(x[x.columns[i]].mode())
        promedio.append(x[x.columns[i]].mean())
        descriptivo.append(x[x.columns[i]].describe())
        sumaacu.append(x[x.columns[i]].cumsum()) 
        maxacu.append(x[x.columns[i]].cummax())
        minacu.append(x[x.columns[i]].cummin())
        productoria.append(x[x.columns[i]].cumprod())
        resultados = pd.DataFrame({'variable':variable, 'media':media, 'mediana':mediana, 'varianza':varianza, 'desviacion':desviacion,
                  'maximo':maximo, 'minimo':minimo, 'conteo':conteo, 'moda':moda, 'promedio':promedio,
                              'descriptivo':descriptivo, 'sumaacu':sumaacu, 'maxacu':maxacu, 'minacu':minacu, 
                                'productoria':productoria})
    return resultados
    
```

```python
resultados = Descriptivo(x = dw)
```
